## ⚙️ Fase 1 — Configuración e importación de librerías

```python
import pandas as pd
from google.colab import files

print('✅ Librerías importadas correctamente')
```

## 📂 Fase 2 — Carga del dataset

Se carga el archivo CSV directamente desde el entorno local usando el uploader de Google Colab.

```python
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

print(f'✅ Dataset cargado: {df.shape[0]} filas, {df.shape[1]} columnas')
df.head()
```

## 🔍 Fase 3 — Exploración inicial

Antes de limpiar, es importante entender la estructura del dato: tipos de columnas, valores nulos y estadísticas básicas.

```python
print('--- Tipos de dato ---')
print(df.dtypes)
print()
print('--- Valores nulos por columna ---')
print(df.isnull().sum())
```

```python
print('--- Estadísticas descriptivas ---')
df.describe()
```

**Hallazgos de la exploración:**
- El dataset no tiene valores nulos — está completo ✅
- Las columnas `date` y `time` están en formato texto y deben convertirse
- Los nombres de columnas son poco descriptivos y contienen caracteres especiales
- No existen columnas derivadas como margen bruto, mes o día de la semana

## 🧹 Fase 4 — Limpieza y transformación

### 4.1 Renombrar columnas

Se estandarizan los nombres a un formato limpio, descriptivo y en snake_case.

```python
df.rename(columns={
    'plk-invoice-number': 'invoice_id',
    'xyz-branch': 'branch',
    'city-ghh-998': 'city',
    'type-usr-search': 'customer_type',
    'category-catalog-dsp': 'category',
    'cost of goods sold': 'cogs',
    'dateddmmyyy': 'date',
    'payment-type-full': 'payment_method'
}, inplace=True)

print('✅ Columnas renombradas')
print(df.columns.tolist())
```

### 4.2 Corregir tipos de dato

Las columnas `date` y `time` estaban como texto. Se convierten a sus tipos correctos para poder trabajar con ellas en análisis temporales.

```python
df['date'] = pd.to_datetime(df['date'])
df['time'] = pd.to_datetime(df['time'], format='%H:%M').dt.time

print('✅ Tipos de dato corregidos')
print(df[['date','time']].dtypes)
```

### 4.3 Crear columnas derivadas

Se generan nuevas variables a partir de las existentes para enriquecer el análisis:
- **gross_margin**: diferencia entre venta y costo
- **month**: nombre del mes extraído de la fecha
- **day_of_week**: día de la semana
- **hour**: hora de la transacción

```python
df['gross_margin'] = df['sale'] - df['cogs']
df['month'] = df['date'].dt.month_name()
df['day_of_week'] = df['date'].dt.day_name()
df['hour'] = pd.to_datetime(df['time'].astype(str), format='%H:%M:%S').dt.hour

print('✅ Columnas derivadas creadas')
df[['date','month','day_of_week','hour','sale','cogs','gross_margin']].head()
```

## ✅ Fase 5 — Validación del dataset limpio

Se verifica que el dataset final esté correcto antes de exportarlo.

```python
print(f'Filas: {df.shape[0]}')
print(f'Columnas: {df.shape[1]}')
print()
print('Tipos de dato finales:')
print(df.dtypes)
print()
print('Nulos:', df.isnull().sum().sum())
df.head()
```

## 💾 Fase 6 — Exportar dataset limpio

```python
df.to_csv('supermarket_clean.csv', index=False)
files.download('supermarket_clean.csv')
print('✅ Archivo exportado: supermarket_clean.csv')
```
